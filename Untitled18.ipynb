```python
import numpy as np
import matplotlib.pyplot as plt

c = 3e8
num_peaks = np.random.randint(10, 25) # Number of Gaussian peaks in the spectrum
frequencies = []
amplitudes = []
gamma_omega_values = []

for _ in range(num_peaks):
    A = 0.1 + np.random.rand() * 0.9
    lambda_0 = np.random.uniform(600, 900) * 10**-9
    gamma_0 = np.random.uniform(150, 220) * 10**-9
    delta_lambda_0_i = (-gamma_0 / 4) + (np.random.rand() * gamma_0) / 2
    lambda_0_i = lambda_0 + delta_lambda_0_i
    omega_0_i = 2 * np.pi * c / lambda_0_i
    gamma_lambda_i = (4 * np.random.rand() + 30) * 10**-9
    gamma_omega_i = 2 * np.pi * c / gamma_lambda_i
    frequencies.append(omega_0_i)
    print(omega_0_i)
    amplitudes.append(A)
    gamma_omega_values.append(gamma_omega_i)
    gaussian_function = A * np.exp((-(4 * np.log(2)) * (omega - omega_0_i)**2) /gamma_omega_i**2)

    plt.plot(omega,gaussian_function, label="Gaussian function")
    plt.xlabel("frequency")
    plt.ylabel("amplitude")
    plt.show()


```

```python
omega_min = min(frequencies) - 5 * max(gamma_omega_values)
omega_max = max(frequencies) + 5 * max(gamma_omega_values)
omega = np.linspace(omega_min, omega_max, 4096)
spectrum = np.zeros_like(omega)

for freq, amp, gamma_omega_i in zip(frequencies, amplitudes, gamma_omega_values):
    spectrum += amp * np.exp((-(4 * np.log(2)) * (omega - freq)**2) / gamma_omega_i**2)

plt.plot(omega, spectrum)
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.title("Spectrum of Gaussian Pulses")
plt.show()
```
